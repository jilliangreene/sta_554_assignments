# HW 7:
In this assignment, we are using a UCI ML Repo dataset to model `alcohol` (MLR) or type of wine (LR). Data downloaded from [here](https://archive.ics.uci.edu/dataset/186/wine+quality)

```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV
from sklearn.linear_model import LinearRegression, LassoCV, Ridge, ElasticNet, LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, log_loss, accuracy_score
```

```python
# Read in red wine csv
reds = pd.read_csv('winequality-red.csv', sep=';')
# Add wine_type col
reds['wine_type'] = 'red'
# print length
print(len(reds))

# White wine
whites = pd.read_csv('winequality-white.csv', sep=';')
whites['wine_type'] = 'white'
print(len(whites))

# Combine the two dfs
df = pd.concat([reds, whites])
print(len(df))
```

Now the two datasets have been read in, added a column for type of wine, and binded together. I've used the `print(len())` tool to ensure the binding happened as expected.

```python
# Use train_test_splt() to do stratified split with equal parts by wine_type
train, test = train_test_split(df, test_size=0.7, stratify=df['wine_type'], random_state=120301)
# Print check
print(len(train[train['wine_type'] == 'white']) / len(test[test['wine_type'] == 'white']))
print(len(train[train['wine_type'] == 'red']) / len(test[test['wine_type'] == 'red']))
```

```python
# Set x and y for train-test split
x_train = train.drop(['alcohol', 'wine_type'], axis=1)
y_train = train['alcohol']

x_test = test.drop(['alcohol', 'wine_type'], axis=1)
y_test = test['alcohol']
```

Now that the data is split up proportionally equal, I can begin the regression task from HW 7, beginning with 4 MLRs

```python
# MLR 1 with interaction terms
model1 = Pipeline([
    ("scaler", StandardScaler()),
    ("interactions", PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)),
    ("lr", LinearRegression())])
```

```python
# MLR 2 (standard)
model2 = Pipeline([("scaler", StandardScaler()),
    ("lr", LinearRegression())])
```

```python
# 2-deg polynomial
model3 = Pipeline([("scaler", StandardScaler()),
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("lr", LinearRegression())])
```

```python
# 4-degree polynomial
model4 = Pipeline([("scaler", StandardScaler()),
    ("poly", PolynomialFeatures(degree=4, include_bias=False)),
    ("lr", LinearRegression())])
```

```python
# Run models and compare CV scores
models = {"Interaction": model1,
    "Standard": model2,
    "Poly2": model3,
    "Poly4": model4}

scoring = {"rmse": "neg_root_mean_squared_error",
    "mae": "neg_mean_absolute_error",
    "r2": "r2"}

results = {}

for name, model in models.items():
    scores = cross_validate(
        model, x_train, y_train,
        cv=5,
        scoring=scoring)

    results[name] = {"RMSE": -np.mean(scores["test_rmse"]),
        "MAE": -np.mean(scores["test_mae"]),
        "R2": np.mean(scores["test_r2"])}

# Print scores
for name, metrics in results.items():
    print(f"\n{name}")
    print(f"RMSE: {metrics['RMSE']:.4f}")
    print(f"MAE:  {metrics['MAE']:.4f}")
    print(f"R2:   {metrics['R2']:.4f}")
```

The best MLR model was the one that used 2-degree polynomial terms (the 4-degree model was significantly bad). The next 3 models require a set of predictors, so I'll filter those down and try the models.

```python
# print x_train col names
x_train.columns
```

```python
# 6-predictor models
x_train6 = x_train[['citric acid', 'residual sugar', 'fixed acidity', 'sulphates', 'pH', 'chlorides']]

x_test6 = x_test[['citric acid', 'residual sugar', 'fixed acidity', 'sulphates', 'pH', 'chlorides']]
```

```python
# LASSO model with > 5 predictors, with CV tuning
lasso_model = Pipeline([("scaler", StandardScaler()),
    ("lasso", LassoCV(cv = 10))])

lasso_model.fit(x_train6, y_train)

best_alpha = lasso_model.named_steps["lasso"].alpha_
print("Best alpha:", best_alpha)

lasso_scores = cross_validate(lasso_model,
    x_train6,
    y_train,
    cv=5,
    scoring=scoring)

lasso_results = {"RMSE": -np.mean(lasso_scores["test_rmse"]),
    "MAE": -np.mean(lasso_scores["test_mae"]),
    "R2":  np.mean(lasso_scores["test_r2"])}

results["LASSO"] = lasso_results

# Print scores
for name, metrics in results.items():
    print(f"\n{name}")
    print(f"RMSE: {metrics['RMSE']:.4f}")
    print(f"MAE:  {metrics['MAE']:.4f}")
    print(f"R2:   {metrics['R2']:.4f}")
```

```python
# Ridge reg. model with > 5 predictors, with CV tuning
ridge_model = Pipeline([("scaler", StandardScaler()),
    ("ridge", Ridge())])

param_grid = {"ridge__alpha": [0.01, 0.1, 1.0, 10.0, 100.0]}

#Ridge is easier to train on 1 scoring stat - will score the real CV on all
ridge_model = GridSearchCV(ridge_model,
    param_grid,
    cv=10,
    scoring="neg_root_mean_squared_error")

ridge_scores = cross_validate(ridge_model,
    x_train6,
    y_train,
    cv=5,
    scoring=scoring)

ridge_results = {"RMSE": -np.mean(ridge_scores["test_rmse"]),
    "MAE": -np.mean(ridge_scores["test_mae"]),
    "R2":  np.mean(ridge_scores["test_r2"])}

results["Ridge"] = ridge_results

# Print scores
for name, metrics in results.items():
    print(f"\n{name}")
    print(f"RMSE: {metrics['RMSE']:.4f}")
    print(f"MAE:  {metrics['MAE']:.4f}")
    print(f"R2:   {metrics['R2']:.4f}")
```

```python
# Elastic Net model with > 5 predictors, with CV tuning
elastic_model = Pipeline([("scaler", StandardScaler()),
    ("elastic", ElasticNet(max_iter=10000))])

param_grid = {"elastic__alpha": [0.01, 0.1, 1.0, 10.0],
    "elastic__l1_ratio": [0.2, 0.5, 0.8]}

elastic_model = GridSearchCV(elastic_model,
    param_grid,
    cv=5,
    scoring="neg_root_mean_squared_error")

elastic_scores = cross_validate(elastic_model,
    x_train6,
    y_train,
    cv=5,
    scoring=scoring)

elastic_results = { "RMSE": -np.mean(elastic_scores["test_rmse"]),
    "MAE": -np.mean(elastic_scores["test_mae"]),
    "R2":  np.mean(elastic_scores["test_r2"])}

results["Elastic Net"] = elastic_results

# Print scores
for name, metrics in results.items():
    print(f"\n{name}")
    print(f"RMSE: {metrics['RMSE']:.4f}")
    print(f"MAE:  {metrics['MAE']:.4f}")
    print(f"R2:   {metrics['R2']:.4f}")
```

```python
# Compare all 4 models on the test data
# Fit models
# model3 = Poly2
model3.fit(x_train, y_train)
lasso_model.fit(x_train6, y_train)
ridge_model.fit(x_train6, y_train)
elastic_model.fit(x_train6, y_train)

# Predict on test
y_pred_poly2 = model3.predict(x_test)
y_pred_lasso = lasso_model.predict(x_test6)
y_pred_ridge = ridge_model.predict(x_test6)
y_pred_elastic = elastic_model.predict(x_test6)

# Print metrics (same as the model dev)
test_results = {}

test_results["Poly2"] = {"RMSE": np.sqrt(mean_squared_error(y_test, y_pred_poly2)),
    "MAE": mean_absolute_error(y_test, y_pred_poly2),
    "R2": r2_score(y_test, y_pred_poly2)}

test_results["LASSO"] = {"RMSE": np.sqrt(mean_squared_error(y_test, y_pred_lasso)),
    "MAE": mean_absolute_error(y_test, y_pred_lasso),
    "R2": r2_score(y_test, y_pred_lasso)}

test_results["Ridge"] = {"RMSE": np.sqrt(mean_squared_error(y_test, y_pred_ridge)),
    "MAE": mean_absolute_error(y_test, y_pred_ridge),
    "R2": r2_score(y_test, y_pred_ridge)}

test_results["Elastic Net"] = {"RMSE": np.sqrt(mean_squared_error(y_test, y_pred_elastic)),
    "MAE": mean_absolute_error(y_test, y_pred_elastic),
    "R2": r2_score(y_test, y_pred_elastic)}

for name, metrics in test_results.items():
    print(f"\n{name}")
    print(f"RMSE: {metrics['RMSE']:.4f}")
    print(f"MAE:  {metrics['MAE']:.4f}")
    print(f"R2:   {metrics['R2']:.4f}")
```

Similar to the training stats, the 2-degree polynomial regression is the best model by far. Interestingly, the stats are better in the testing than the training for the `Poly2` model. This could be a result of a small CV, perhaps weighted by a bad model.

Now we will do a classification model test using log regression models to predict `wine_type`.

```python
# Set x and y for train-test split
x_train_class = train.drop(['wine_type'], axis=1)
y_train_class = train['wine_type']

x_test_class = test.drop(['wine_type'], axis=1)
y_test_class = test['wine_type']
```

```python
# Logistic reg models
log_base = Pipeline([("scaler", StandardScaler()),
    ("log", LogisticRegression(penalty=None, max_iter=10000))])

# Ridge
log_l2 = Pipeline([("scaler", StandardScaler()),
    ("log", LogisticRegression(penalty="l2", max_iter=10000))])

param_grid_l2 = {"log__C": [0.01, 0.1, 1.0, 10.0]}

# LASSO
log_l1 = Pipeline([("scaler", StandardScaler()),
    ("log", LogisticRegression(penalty="l1", solver="liblinear", max_iter=10000))])

param_grid_l1 = {"log__C": [0.01, 0.1, 1.0, 10.0]}

# Elastic
log_en = Pipeline([("scaler", StandardScaler()),
    ("log", LogisticRegression(penalty="elasticnet",
        solver="saga",
        max_iter=10000))])

param_grid_en = {"log__C": [0.01, 0.1, 1.0], "log__l1_ratio": [0.2, 0.5, 0.8]}
```

```python
# Run parameter tuning on Ridge, LASSO, and Elastic models using negative
# log loss as metric
log_l2 = GridSearchCV(log_l2, param_grid_l2, cv=5, scoring="neg_log_loss")
log_l1 = GridSearchCV(log_l1, param_grid_l1, cv=5, scoring="neg_log_loss")
log_en = GridSearchCV(log_en, param_grid_en, cv=5, scoring="neg_log_loss")
```

```python
# Train models
models = {"Baseline": log_base,
    "Ridge": log_l2,
    "LASSO": log_l1,
    "Elastic Net": log_en}

cv_results = {}

for name, model in models.items():
    scores = cross_validate(model,
        x_train_class,
        y_train_class,
        cv=5,
        scoring="neg_log_loss")

    cv_results[name] = -np.mean(scores["test_score"])

for name, ll in cv_results.items():
    print(f"{name}: Log-Loss = {ll:.4f}")
```

In training, it looks like LASSO is the best model by a very small margin. Now I'll do the testing on the held out data.

```python
# Fit models
for model in models.values():
    model.fit(x_train_class, y_train_class)

# Testing
test_results = {}

for name, model in models.items():
    # Probabilities for log-loss
    y_prob = model.predict_proba(x_test_class)

    # Class predictions for accuracy
    y_pred = model.predict(x_test_class)

    test_results[name] = {"Log-Loss": log_loss(y_test_class, y_prob),
        "Accuracy": accuracy_score(y_test_class, y_pred)}

for name, metrics in test_results.items():
    print(f"\n{name}")
    print(f"Log-Loss: {metrics['Log-Loss']:.4f}")
    print(f"Accuracy: {metrics['Accuracy']:.4f}")
```

Again, the LASSO model has the lowest log-loss; however the accuracy between all the models is statistically identical. In this case, I would say that any of the models are going to excel at classifying wine type based on the available data!

# HW 8

This time will will be predicting `quality` rather than `alcohol`.

```python
# Import packages for tree modeling
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_validate
```

```python
# Set x and y for train-test split
x_train_trees = train.drop(['quality', 'wine_type'], axis=1)
y_train_trees = train['quality']

x_test_trees = test.drop(['quality', 'wine_type'], axis=1)
y_test_trees = test['quality']
```

```python
# Initialize model
tree_model = DecisionTreeRegressor(random_state=120301)

# Tuning grid
param_grid = {"max_depth": [3, 5, 7, 10, None],
    "min_samples_leaf": [1, 5, 10, 20]}

# CV
tree_model = GridSearchCV(tree_model,
    param_grid,
    cv=5,
    scoring="neg_root_mean_squared_error")

tree_scores = cross_validate(tree_model,
    x_train_trees,
    y_train_trees,
    cv=5,
    scoring=scoring,
    return_estimator=True)

tree_results = {"RMSE": -np.mean(tree_scores["test_rmse"]),
    "MAE": -np.mean(tree_scores["test_mae"]),
    "R2":  np.mean(tree_scores["test_r2"])}

print("\nDecision Tree")
print(f"RMSE: {tree_results['RMSE']:.4f}")
print(f"MAE:  {tree_results['MAE']:.4f}")
print(f"R2:   {tree_results['R2']:.4f}")
```

```python
# Initialize model
rf_model = RandomForestRegressor(random_state=120301)

# Tuning grid
param_grid = {"max_features": [2, 3, 5, 8]}

# CV
rf_model = GridSearchCV(rf_model,
    param_grid,
    cv=5,
    scoring="neg_root_mean_squared_error")

rf_scores = cross_validate(rf_model,
    x_train_trees,
    y_train_trees,
    cv=5,
    scoring=scoring,
    return_estimator=True)

rf_results = {"RMSE": -np.mean(rf_scores["test_rmse"]),
    "MAE": -np.mean(rf_scores["test_mae"]),
    "R2":  np.mean(rf_scores["test_r2"])}

print("\nRandom Forest")
print(f"RMSE: {rf_results['RMSE']:.4f}")
print(f"MAE:  {rf_results['MAE']:.4f}")
print(f"R2:   {rf_results['R2']:.4f}")
```

CV scores look much worse than the MLR models. Let's run on the testing dataset

```python
# Run tree_model on testing data
tree_model.fit(x_train_trees, y_train_trees)
y_pred_tree = tree_model.predict(x_test_trees)

# Random Forest
rf_model.fit(x_train_trees, y_train_trees)
y_pred_rf = rf_model.predict(x_test_trees)

# Calculate RMSE and MAE (and R2 for fun)
# Print metrics (same as the model dev)
test_results = {}

test_results["Tree"] = {"RMSE": np.sqrt(mean_squared_error(y_test_trees, y_pred_tree)),
    "MAE": mean_absolute_error(y_test_trees, y_pred_tree),
    "R2": r2_score(y_test_trees, y_pred_tree)}

test_results["RF"] = {"RMSE": np.sqrt(mean_squared_error(y_test_trees, y_pred_rf)),
    "MAE": mean_absolute_error(y_test_trees, y_pred_rf),
    "R2": r2_score(y_test_trees, y_pred_rf)}

for name, metrics in test_results.items():
    print(f"\n{name}")
    print(f"RMSE: {metrics['RMSE']:.4f}")
    print(f"MAE:  {metrics['MAE']:.4f}")
    print(f"R2:   {metrics['R2']:.4f}")
```

The scores again are looking a lot worse than the MLR models; however, we are predicting `quality` rather than `alcohol`, so it isn't surprising that it's harder to glean the `quality` from the chemical properties than it is the `alcohol` content. Because of this, it is also difficult to compare the MAE and RMSE because it's different units and range. Traditionally, decision tree models are much better performers than linear regressions.

# Classification Tree Models

Here, we can reuse the train and testing objects from HW7, as we are predicting the same variable.

```python
# Initialize model
tree_model_class = DecisionTreeClassifier(random_state=120301)

# new scoring dict
scoring_class = {"log_loss": "neg_log_loss",
    "accuracy": "accuracy"}

# Tuning grid
param_grid = {"max_depth": [3, 5, 7, 10, None],
    "min_samples_leaf": [1, 5, 10, 20]}

# CV
tree_model_class = GridSearchCV(tree_model_class,
    param_grid,
    cv=5,
    scoring="neg_log_loss")

tree_scores = cross_validate(tree_model_class,
    x_train_class,
    y_train_class,
    cv=5,
    scoring=scoring_class,
    return_estimator=True)

tree_results = {"Log-Loss": -np.mean(tree_scores["test_log_loss"]),
    "Accuracy": np.mean(tree_scores["test_accuracy"])}

# Print results
print("\nDecision Tree Classifier")
print(f"Log-Loss: {tree_results['Log-Loss']:.4f}")
print(f"Accuracy: {tree_results['Accuracy']:.4f}")
```

```python
# RF Classification
# Initialize model
rf_model_class = RandomForestClassifier(random_state=120301)

# Tuning grid
param_grid = {"max_depth": [3, 5, 7, 10, None],
    "min_samples_leaf": [1, 5, 10, 20]}

# CV
rf_model_class = GridSearchCV(rf_model_class,
    param_grid,
    cv=5,
    scoring="neg_log_loss")

rf_scores = cross_validate(rf_model_class,
    x_train_class,
    y_train_class,
    cv=5,
    scoring=scoring_class,
    return_estimator=True)

rf_results = {"Log-Loss": -np.mean(rf_scores["test_log_loss"]),
    "Accuracy": np.mean(rf_scores["test_accuracy"])}

# Print results
print("\nRF Classifier")
print(f"Log-Loss: {rf_results['Log-Loss']:.4f}")
print(f"Accuracy: {rf_results['Accuracy']:.4f}")
```

```python
# Testing classifiers
tree_model_class.fit(x_train_class, y_train_class)
y_pred_tree = tree_model_class.predict(x_test_class)
y_prob_tree = tree_model_class.predict_proba(x_test_class)

# Random Forest
rf_model_class.fit(x_train_class, y_train_class)
y_pred_rf = rf_model_class.predict(x_test_class)
y_prob_rf = rf_model_class.predict_proba(x_test_class)

# Calculate RMSE and MAE (and R2 for fun)
# Print metrics (same as the model dev)
test_results = {}

test_results["Tree"] = {"Log-Loss": log_loss(y_test_class, y_prob_tree),
    "Accuracy": accuracy_score(y_test_class, y_pred_tree)}

test_results["RF"] = {"Log-Loss": log_loss(y_test_class, y_prob_rf),
    "Accuracy": accuracy_score(y_test_class, y_pred_rf)}

for name, metrics in test_results.items():
    print(f"\n{name}")
    print(f"Log-Loss: {metrics['Log-Loss']:.4f}")
    print(f"Accuracy:  {metrics['Accuracy']:.4f}")
```

The classifier models compare very well to the previous ones, as they are predicting the same thing (`wine_type`) and their performance is nearly identical. Again, this is likely because it is much easier to discern either `alcohol` content or `wine_type` from the chemical predictor variables, whereas `quality` can be a bit more subjective based on other factors. Because of this, the decision trees performed worse in regressions but highly similar in classifications.
